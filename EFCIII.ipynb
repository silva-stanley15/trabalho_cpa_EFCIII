## Problema da mochila 

```python
import random
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gerarVetorV(n):
  A = []
  for i in range(n):
    A.append(random.randrange(10, 300, 10))
  return A

def gerarVetorP(n):
  A = []
  for i in range(n):
    A.append(random.randrange(10, 100, 10))
  return A
```

```python
def mochila(capacidade, pesos, valores, n):
  ini= time.time()
  vetor = [[0 for x in range(capacidade + 1)] for x in range(n + 1)]

  for i in range(n + 1):
      for w in range(capacidade + 1):
          if i == 0 or w == 0:
              vetor[i][w] = 0
          elif pesos[i-1] <= w:
              vetor[i][w] = max(valores[i-1] + vetor[i-1][w-pesos[i-1]],  vetor[i-1][w])
          else:
              vetor[i][w] = vetor[i-1][w]
  fim = time.time()
  tempo_exec=fim-ini
  return tempo_exec
```

```python
aux = []
numEntradas = []

for i in range (5, 30):
  numEntradas.append(i)
  valores = gerarVetorV(i)
  pesos = gerarVetorP(i)
  capacidade = 50
  aux.append(mochila(capacidade, pesos, valores, i))


plt.plot(numEntradas,aux, 'r', label='subsetsum')
plt.xlabel("qtde entradas")
plt.ylabel("tempo")
plt.legend()
plt.show() 
```
